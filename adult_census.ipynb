```python
df.drop(columns=['race'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df.drop(columns=['native.country'],axis=1,inplace=True)
df.shape
```

```python
df=pd.get_dummies(data=df,columns=['workclass','education','marital.status','occupation','relationship','sex','income'],drop_first=True)
df.shape
```

```python
df.head(2)
```

```python
x=df.drop(columns=['income_>50K'],axis=1)
y=df['income_>50K'].values.reshape(-1,1)
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
mod1=LogisticRegression()
mod1.fit(x_train,y_train)
```

```python
print("\n Training Score",mod1.score(x_train,y_train))
print("\n Testing Score",mod1.score(x_test,y_test))
```

```python
y_pred1=mod1.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report
cm1=confusion_matrix(y_test,y_pred1)
sns.heatmap(cm1,annot=True)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
mod2=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
mod2.fit(x_train,y_train)
```

```python
print("\n Training Score",mod2.score(x_train,y_train))
print("\n Testing Score",mod2.score(x_test,y_test))
```

```python
y_pred2=mod2.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report
cm2=confusion_matrix(y_test,y_pred2)
sns.heatmap(cm2,annot=True)
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier
mod3=DecisionTreeClassifier()
mod3.fit(x_train,y_train)
```

```python
print("\n Training Score",mod3.score(x_train,y_train))
print("\n Testing Score",mod3.score(x_test,y_test))
```

```python
y_pred3=mod3.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report
cm3=confusion_matrix(y_test,y_pred3)
sns.heatmap(cm3,annot=True)
```

```python
print(classification_report(y_test,y_pred3))
```

```python
from sklearn.ensemble import RandomForestClassifier
mod4=RandomForestClassifier()
mod4.fit(x_train,y_train)
```

```python
print("\n Training Score",mod4.score(x_train,y_train))
print("\n Testing Score",mod4.score(x_test,y_test))
```

```python
y_pred4=mod4.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report
cm4=confusion_matrix(y_test,y_pred4)
sns.heatmap(cm4,annot=True)
```

```python
print(classification_report(y_test,y_pred4))
```

```python
from sklearn.naive_bayes import GaussianNB
mod5=GaussianNB()
mod5.fit(x_train,y_train)
```

```python
y_pred5=mod5.predict(x_test)
from sklearn.metrics import confusion_matrix,classification_report
cm5=confusion_matrix(y_test,y_pred5)
sns.heatmap(cm5,annot=True)
```

```python
print("\n Training Score",mod5.score(x_train,y_train))
print("\n Testing Score",mod5.score(x_test,y_test))
```

```python
print(classification_report(y_test,y_pred5))
```

```python
labels=['Logistics Regression','KNN','Decision Tree','Random Forest','Naive-Bayes']
comparison_df=pd.DataFrame(columns=['Training','Testing'],index=labels)
training=[mod1.score(x_train,y_train),mod2.score(x_train,y_train),mod3.score(x_train,y_train),mod4.score(x_train,y_train),mod5.score(x_train,y_train)]
testing=[mod1.score(x_test,y_test),mod2.score(x_test,y_test),mod3.score(x_test,y_test),mod4.score(x_test,y_test),mod5.score(x_test,y_test)]
comparison_df['Training']=training
comparison_df['Testing']=testing
comparison_df
```

```python
plt.figure(figsize=(8,5))
plt.title('Training Set')
sns.barplot(labels,training)
```

```python
plt.figure(figsize=(8,5))
plt.title('Testing Set')
sns.barplot(labels,testing)
```
